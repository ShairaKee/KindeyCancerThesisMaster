# Dataset Organization
In this section we will create folders and subfolders which will be where the tiled slides will be placed. As well as, filtering out the datasets which have genemoic and image data which was manually placed in a excel file converted into a csv. These files will the be moved to their respective folders and the files without image data and genemoic data will be exculded from the dataset used.

#### Importing Dataset

```python
import csv
import os
import shutil
```

#### Importing the CSV file with the list of datas that have genomic and image data
Firstly the csv file which had the image and genomic data were imported which will be used as a filter later on. Second step was converting the csv data into a list to put the data into rows, then it was converted into tuple to strip/get rid of the /n and Patient-ID row. Note: this could have been easier done with pandas.

```python
# Importing CSV files with the names used
file = open('C:/Users/aaron/Desktop/NewName.csv') #loading in the data
csvreader = csv.reader(file) # reading the csv
```

```python
# Converting the CSV to list
rows = [] #converting to list
for row in file:
    rows.append(row)

rows
```

```python
# Converting to Tuple
x = [s.strip('\n') for s in rows] # stripping/removing \n
x2 = [s.strip('Patient-ID') for s in x] #removing Patient ID
y = tuple(x2) #Converting to tuple
print(y)
```

#### Copying all files that matches the excel file names
This step is to segregate all the svs files that are found in the list from the files that are no in the list above

```python
src_dir = 'D:/Oimages/' #Source Library which contains the svs files (complete dataset)
dest_dir = 'D:/level0/' #Destination Library where you want to move the svs files

for (dirpath, dirnames, filenames) in os.walk(src_dir): #Looping over the source library directories and sub-directories and saving filename,directorynames and directory paths
    for fname in filenames: # placing the filenames from the folder into the fname variable
        if fname.startswith(y) and fname.endswith(".svs"):# conditional statement that looks at the list generated and matches it with the filenames from the folder
            shutil.copy(os.path.join(dirpath, fname),dest_dir ) # Copying the files that have the same filenames into destination folder
```

#### Creating folders
This step will create folders based on the filenames of the svs files this will be where the tiled data will be stored which will be in section 02

```python
new_dic='C:/Users/aaron/Desktop/level0/'
newer = 'D:/Level1/'
for file in os.walk(new_dic): # for loop which iterates over svs files and gets its filenames
    for fname in file: # iterates over file names
        os.mkdir(os.path.join(newer,fname)) # creates a folder with the name of the file as the folder name
```

### Note: Insight to Folders Names 
* Oimage - Contains all the original SVS files
* level0 - Will contain the SVS files that will be used (Main Dataset Folder)
* level1 - Contain folders with the SVS filenames from level0 will be used in tiling later
